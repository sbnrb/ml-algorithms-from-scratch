```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import random
n = 75

c0 = round(random.random() * 25, 2)
c1 = round(random.random() * 5, 2)
```

```python
x = np.arange(0, n, 1).reshape((n, 1))
y = c0 + c1*x + np.random.normal(0, 25, x.shape)
```

```python
plt.figure(figsize = (15,7))
plt.plot(x, y, 'co')
plt.plot(x, c0 + c1*x , 'm-', label = 'true value')
```

```python
X = np.ones((n, 1))
X = np.append(X, x, axis=1)
```

```python
c_pred = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)), X.T), y)
```

```python
c_pred
```

```python
plt.figure(figsize = (15,7))
plt.plot(x, y, 'co')
line1, = plt.plot(x, c0 + c1*x, 'm-', label = 'true value')
line2, = plt.plot(x, c_pred[0] + c_pred[1]*x, 'r-', label = 'predicted value')

legend = plt.legend(handles=[line1, line2])
```

```python
print(f'real: c0 = {c0}, c1 = {c1}')
print(f'predicted: c0 = {c_pred[0]}, c1 = {c_pred[1]}')
```

```python
from sklearn.linear_model import LinearRegression

clf = LinearRegression(fit_intercept = False)
clf.fit(X[:], y[:])

clf.coef_
```
