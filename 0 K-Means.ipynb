```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import random
```

```python
x1 = np.random.normal(3, 1.2, 350)
y1 = np.random.normal(2, 1.2, 350)

x2 = np.random.normal(7, 1.2, 350)
y2 = np.random.normal(4, 1.2, 350)

x3 = np.random.normal(6, 1, 300)
y3 = np.random.normal(0, 1, 300)

x = np.append(np.append(x1, x2), x3)
y = np.append(np.append(y1, y2), y3)
```

```python
plt.figure(figsize = (9, 6))

plt.plot(x, y, '.', color = 'gray')
```

```python
def k_means(k = 3):
    i_centr = np.arange(0, len(x), 1)
    random.shuffle(i_centr)
    i_centr = i_centr[:k]
    centr_x, centr_y = x[i_centr], y[i_centr]

    cont = True
    while cont:    
        plt.figure(figsize = (8, 5))

        clusters = []
        for j in range(len(x)):
            dist, new_centr = [], []

            for i in range(k):
                dist += [np.linalg.norm((x[j] - centr_x[i], y[j] - centr_y[i]))]
                plt.plot(centr_x[i], centr_y[i], '.', color = 'C' + str(i))

            clusters += [np.argmin(dist)]
            plt.plot(x[j], y[j], '.', color = 'C' + str(clusters[-1]), alpha = 0.4)

        new_centr_x, new_centr_y = [], []

        cont = False
        for i in range(k):
            centr_x_i, centr_y_i, N = 0, 0, 0

            for j in range(len(x)):
                if clusters[j] == i:
                    centr_x_i += x[j]
                    centr_y_i += y[j]
                    N += 1

            new_centr_x += [centr_x_i / N]
            new_centr_y += [centr_y_i / N]

            dx = new_centr_x[i] - centr_x[i]
            dy = new_centr_y[i] - centr_y[i]

            if dx != 0 or dy != 0:
                cont = True

            plt.plot(new_centr_x[i], new_centr_y[i], '.', color = 'C' + str(i))
            plt.arrow(centr_x[i], centr_y[i], dx, dy, 
                      shape = 'full', color = 'C' + str(i), head_width = 0.2, width = 0.02, length_includes_head = True)

        centr_x, centr_y = new_centr_x, new_centr_y
        
k_means()
```

```python
k_means(k = 2)
```

```python
k_means(k = 4)
```
